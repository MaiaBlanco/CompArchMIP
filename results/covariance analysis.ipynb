```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing

files1 = ["smart_64x64_CACHE.csv", "64x64_CACHE_SMART_2.csv"]
files2 = ["64x64_SPAD.csv"] #, "64x64_SPAD_2_block.csv", "64x64_SPAD_3_cyclic.csv"]
outputs = ["Total Area", "Avg Power", "Runtime"]

files=files2
dfs = []

for f in files:
    dfs.append(pd.read_csv(f))

df = pd.concat(dfs)
df.reset_index(drop=True, inplace=True)
df.drop(['arrays','memory_type'], axis=1, inplace=True)
df['Runtime'] = (df['Cycle '] * df.cycle_time).values / 1000 # Runtime in us

if 'unrolling_factor_norm' in df.columns and 'unrolling_factor_cols' in df.columns:
    df['unroll_norm_cols'] = df['unrolling_factor_norm'].combine_first(df['unrolling_factor_cols'])
    df.drop(['unrolling_factor_norm'], axis=1, inplace=True)
if 'unrolling_factor_sub' in df.columns and 'unrolling_factor_cols' in df.columns:
    df['unroll_sub_cols'] = df['unrolling_factor_sub'].combine_first(df['unrolling_factor_cols'])
    df.drop(['unrolling_factor_cols'], axis=1, inplace=True)
    df.drop(['unrolling_factor_sub'], axis=1, inplace=True)
del_cols = ['Avg FU Dynamic Power', 'Avg FU Power', 'Avg FU leakage Power',
       'Avg MEM Dynamic Power', 'Avg MEM Leakage Power', 'Avg MEM Power',
       'Avg Power', 'Cycle ', 'FU Area', 'Idle FU Cycles', 'MEM Area',
       'Num of Registers (32-bit)', 'Num of Shifters (32-bit)',
       'Num of Single Precision FP Adders',
       'Num of Single Precision FP Multipliers', 'Running ',
       'Top level function', 'Total Area', 'Runtime', 'Unnamed: 0',]
o_cols = []
for c in del_cols:
    if c in df.columns:
        o_cols.append(c)
    
```
